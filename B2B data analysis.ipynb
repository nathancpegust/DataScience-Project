```python
import pandas as pd
import os
```

```python
df = pd.read_excel("Gross margin.xlsx",index_col=[0])
df.head()
```

# Q1: Find Inventory turnover ratio

```python
df.sort_values(by=['Inventory Turnover Ratio'],ascending=False)
```

# Q2: All Inventory which makes us lose money


```python

Losingmoneyreport=df.loc[df["Gross Margin"]< '0']

Losingmoneyreport.sort_values(by=['Gross Margin'],ascending=False)
```

# Q3: Find all Dead inventories

```python
df = pd.read_excel("inventory.xlsx",index_col=[0])
df.head()
```

```python
deadinventoryreport=df.loc[df["Qty"]!= '0.0']

deadinventoryreport.sort_values(by=['Qty'],ascending=False)
```
